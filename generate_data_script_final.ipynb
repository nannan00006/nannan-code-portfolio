```python
# impprt necessary libraries
import pandas as pd
import numpy as np
import faker
import random
from datetime import datetime, timedelta
```

```python

fake = faker.Faker()

# Set the random seed to ensure the reproducibility of the results
random.seed(42)
np.random.seed(42)

# Functions for generating random data
def generate_offices(n):
    offices = []
    for i in range(n):
        offices.append({
            "office_id": i + 1,
            "address": fake.street_address(),
            "city": fake.city(),
            "state": random.choice(["NY", "NJ", "CT"]),
        })
    return offices

# To make the data appear more realistic, we have capped the monthly income of employees between 2000 and 10,000 US dollars, 
# and randomly assigned these employees to hypothetical 10 offices.
def generate_employees(n):
    employees = []
    for i in range(n):
        employees.append({
            "employee_id": i + 1,
            "first_name": fake.first_name(),
            "last_name": fake.last_name(),
            "email": fake.email(),
            "contact_number": fake.numerify(text='###-###-####'),
            "address": fake.street_address(),
            "agent": random.choice(["Yes", "No"]),
            "employment_status": random.choice(["full-time", "part-time"]),
            "education": random.choice(["High School", "Bachelor's", "Master's", "PhD"]),
            "monthly_salary": round(random.uniform(2000, 10000), 2),
            "position": random.choice(["Manager", "Others", "Director"]),
            "office_id": random.randint(1, 10),
        })
    return employees

# We set the number of bedrooms in the housing assets to range between 1 and 5, the number of bathrooms between 1 and 3, 
# the area between 500 and 1000 square feet, and the construction year of the houses between 1 and 30 years, to make the data as similar as possible to real-world data.
def generate_properties(n):
    properties = []
    for i in range(n):
        properties.append({
            "property_id": i + 1,
            "address": fake.street_address(),
            "city": fake.city(),
            "state": random.choice(["NY", "NJ", "CT"]),
            "bedroom_num": random.randint(1, 5),
            "bathroom_num": random.randint(1, 3),
            "sqft": round(random.uniform(500, 3000), 2),
            "floor": random.randint(1, 10),
            "elevator": random.choice(["Yes", "No"]),
            "washer_dryer": random.choice(["Yes", "No"]),
            "air_conditioner": random.choice(["Yes", "No"]),
            "age": round(random.uniform(1, 30), 1),
        })
    return properties


def generate_property_owners(n):
    owners = []
    for i in range(n):
        owners.append({
            "owner_id": i + 1,
            "first_name": fake.first_name(),
            "last_name": fake.last_name(),
            "email": fake.email(),
            "contact_number": fake.numerify(text='###-###-####'),
            "current_address": fake.street_address(),
        })
    return owners

# We assume that all historical listings cover a total of 10,000 different properties, these assets are owned by 5,000 owners, 
# and have been managed by 800 employees. The listing prices for the houses are between 100,000 and 500,000 US dollars, 
# and the types of listings are for rent or for sale.

def generate_listings(n):
    listings = []
    for i in range(n):
        listings.append({
            "listing_id": i + 1,
            "property_id": random.randint(1, 10000),
            "owner_id": random.randint(1, 5000),
            "employee_id": random.randint(1, 800),
            "listing_date": fake.date_this_decade(),
            "listing_price": round(random.uniform(100000, 500000), 2),
            "listing_type": random.choice(["sell", "rent"]),
            "status": random.choice(["Available", "Pending", "Not Available"]),
        })
    return listings


def generate_clients(n):
    clients = []
    for i in range(n):
        clients.append({
            "client_id": i + 1,
            "first_name": fake.first_name(),
            "last_name": fake.last_name(),
            "email": fake.email(),
            "contact_number": fake.numerify(text='###-###-####'),
            "current_address": fake.street_address(),
        })
    return clients

# We assume that in the "client_preferences" table, we have collected housing preference information for a total of 8,000 clients (including those who buy or rent houses). 
# The table also covers their minimum and maximum budgets, making it convenient for the company's agents to recommend the most suitable housing information to clients.

def generate_client_preferences(n):
    preferences = []
    used_client_ids = set()
    
    while len(preferences) < n:
        new_client_id = random.randint(1, 8000)
        
        if new_client_id not in used_client_ids:
            dealing_type = random.choice(["buy", "rent"])
            
            if dealing_type == "rent":
                min_budget = round(random.uniform(100, 500), 2)
                max_budget = round(random.uniform(500, 15000), 2)
            else:
                min_budget = round(random.uniform(50000, 200000), 2)
                max_budget = round(random.uniform(200000, 500000), 2)
                
            preferences.append({
                "client_id": new_client_id,
                "preferred_location": fake.city(),
                "min_budget": min_budget,
                "max_budget": max_budget,
                "dealing_type": dealing_type,
                "property_type": random.choice(["Single Family", "Apartment", "Condo"]),
                "bedrooms": random.randint(1, 5),
                "bathrooms": random.randint(1, 3),
            })
            used_client_ids.add(new_client_id)
    
    return preferences

# We assume that among the 30,000 listing entries, 20,000 are rental contracts, involving a total of 5,000 clients.

def generate_rental_leases(n):
    leases = []
    for i in range(n):
        start_date = fake.date_this_year()
        end_date = start_date + timedelta(days=365)
        leases.append({
            "lease_id": i + 1,
            "listing_id": random.randint(1, 20000),
            "client_id": random.randint(1, 5000),
            "lease_start": start_date,
            "lease_end": end_date,
        })
    return leases

# We assume that among the 30,000 listings, 10,000 belong to purchase contracts, involving a total of 5,000 customers (to distinguish between rentals and purchases, 
# the listing IDs for purchases are between 20,000 and 30,000, ensuring there is no overlap with rentals).

def generate_sale_contracts(n):
    contracts = []
    for i in range(n):
        contracts.append({
            "contract_id": i + 1,
            "listing_id": random.randint(20000, 30000),
            "client_id": random.randint(5001,8000),
            "signing_date": fake.date_this_year(),
        })
    return contracts

# We assume that all 10000 properties have open houses events

def generate_open_houses(n):
    open_houses = []
    for i in range(n):
        open_houses.append({
            "openhouse_id": i + 1,
            "property_id": random.randint(1,10000),
            "date": fake.date_this_year(),
            "start_time": fake.time(),
            "end_time": fake.time(),
        })
    return open_houses

# We assume that in the appointments table, it involves all 10,000 properties, 8,000 clients, and 800 employees, recording the viewing appointments for each property.

def generate_appointments(n):
    appointments = []
    for i in range(n):
        appointments.append({
            "appointment_id": i + 1,
            "client_id": random.randint(1, 8000),
            "property_id": random.randint(1, 10000),
            "employee_id": random.randint(1, 800),
            "appointment_date": fake.date_time_this_year(),
            "status": random.choice(["Scheduled", "Completed", "Canceled"]),
        })
    return appointments

# This table contains information about schools, including their ratings, laying the foundation for subsequent information about schools in the vicinity of the properties.

def generate_schools(n):
    schools = []
    for i in range(n):
        schools.append({
            "school_id": i + 1,
            "school_name": fake.company(),
            "address": fake.street_address(),
            "rating": round(random.uniform(1, 5), 1),
        })
    return schools

# We assume there are a total of 100 well-known schools. This table shows the housing information within 2000 meters of these schools and the distance between them.

def generate_school_property(n):
    school_property = []
    used_combinations = set()  
    
    while len(school_property) < n:
        new_school_id = random.randint(1, 100)
        new_property_id = random.randint(1, 10000)
        
        combination = (new_school_id, new_property_id)
        
        if combination not in used_combinations:
            school_property.append({
                "school_id": new_school_id,
                "property_id": new_property_id,
                "distance": random.randint(100, 2000),
            })
            used_combinations.add(combination) 
    
    return school_property

# This table contains information about amenities, including their types, laying the foundation for subsequent information about amenities in the vicinity of the properties.

def generate_amenities(n):
    amenities = []
    for i in range(n):
        amenities.append({
            "amenity_id": i + 1,
            "amenity_type": random.choice(["Park", "Gym", "Pool"]),
        })
    return amenities

# We assume there are a total of 1000 amenities. This table shows the housing information within 2000 meters of these amenities and the distance between them. The code
# also ensures that the composite primary key is unique.

def generate_amenity_property(n):
    amenity_property = []
    used_combinations = set()  
    
    while len(amenity_property) < n:
        new_amenity_id = random.randint(1, 1000)
        new_property_id = random.randint(1, 10000)
        
        combination = (new_amenity_id, new_property_id)
        
        if combination not in used_combinations:
            amenity_property.append({
                "amenity_id": new_amenity_id,
                "property_id": new_property_id,
                "distance": random.randint(100, 2000),
            })
            used_combinations.add(combination)  
    
    return amenity_property

# This table helps to record the commission information of employees corresponding to rental orders.

def generate_rental_commissions(n):
    commissions = []
    for i in range(n):
        commissions.append({
            "rental_commission_id": i + 1,
            "lease_id": random.randint(1, 20000),
            "commission_amount": round(random.uniform(100, 1000), 2),
            "date_issued": fake.date_this_year(),
            "status": random.choice(["Pending", "Paid"]),
        })
    return commissions

# This table helps to record the commission information of employees corresponding to purchase contracts.

def generate_sale_commissions(n):
    commissions = []
    for i in range(n):
        commissions.append({
            "commission_id": i + 1,
            "contract_id": random.randint(1, 10000),
            "commission_amount": round(random.uniform(1000, 10000), 2),
            "date_issued": fake.date_this_year(),
            "status": random.choice(["Pending", "Paid"]),
        })
    return commissions


# This table displays the total expenditure of 10 offices, including budget and actual expenditure, facilitating the enterprise to track expenditure status.

def generate_office_expenses(n):
    expenses = []
    for i in range(n):
        expenses.append({
            "expense_id": i + 1,
            "office_id": random.randint(1, 10),
            "expense_type": random.choice(["Utilities", "Maintenance", "Office Supplies"]),
            "budget_allocation": round(random.uniform(500, 5000), 2),
            "actual_spending": round(random.uniform(500, 5000), 2),
            "expense_date": fake.date_this_year(),
        })
    return expenses

# This table presents the revenue situation of the ten offices. If the revenue type is 'sale', then only the `sale_income_amount` and `contract_id` 
# columns are non-empty, ensuring the authenticity and accuracy of the data.


def generate_total_income(n):
    incomes = []
    for i in range(n):
        income_type = random.choice(["sale", "rent", "other"])

        income = {
            "income_id": i + 1,
            "office_id": random.randint(1, 10),
            "income_date": fake.date_this_year(),
            "income_type": income_type,
            "sale_income_amount": None,
            "contract_id": None,
            "rent_income_amount": None,
            "lease_id": None,
            "other_income_amount": None
        }

        if income_type == "sale":
            income["sale_income_amount"] = round(random.uniform(50000, 500000), 2)
            income["contract_id"] = random.randint(1, 10000)
        elif income_type == "rent":
            income["rent_income_amount"] = round(random.uniform(500, 5000), 2)
            income["lease_id"] = random.randint(1, 20000)
        elif income_type == "other":
            income["other_income_amount"] = round(random.uniform(500, 5000), 2)

        incomes.append(income)
    return incomes



# generate sample data
data = {
    "offices": generate_offices(10),
    "employees": generate_employees(1000),
    "properties": generate_properties(10000),
    "property_owners": generate_property_owners(5000),
    "listings": generate_listings(30000),
    "clients": generate_clients(8000),
    "client_preferences": generate_client_preferences(5000),
    "rental_leases": generate_rental_leases(20000),
    "sale_contracts": generate_sale_contracts(10000),
    "open_houses": generate_open_houses(10000),
    "appointments": generate_appointments(20000),
    "schools": generate_schools(100),
    "school_property": generate_school_property(1000),
    "amenities": generate_amenities(1000),
    "amenity_property": generate_amenity_property(1000),
    "rental_commissions": generate_rental_commissions(20000),
    "sale_commissions": generate_sale_commissions(10000),
    "office_expenses": generate_office_expenses(1000),
    "total_income": generate_total_income(30000),
}

# generate the dataframe
dfs = {table: pd.DataFrame(records) for table, records in data.items()}

# write the dataframe into csv file
with pd.ExcelWriter("/Users/angelhuang/Desktop/modified_dataset_sample.xlsx") as writer:
    for table, df in dfs.items():
        df.to_excel(writer, sheet_name=table, index=False)
```
